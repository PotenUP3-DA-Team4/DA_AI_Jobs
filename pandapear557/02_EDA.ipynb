```python
import pandas as pd

# comments (254637, 5)
df = pd.read_csv('../chwonjae/data/youtube_comments.csv')

# meta_df
meta_df = pd.read_csv('../chwonjae/data/youtube_video_meta.csv')

vid_list_df = pd.read_csv('../data/channel_videos_with_stats.csv')

print(df.columns)
print(meta_df.columns)
print(vid_list_df.columns)
```

```python

# series to dic key : video_id, val : video_title
# header : Showcase, Most_viewed, Recent_released
# group :  플레이브, 아이브, 이세돌, 라이브
meta_dict = meta_df.set_index('video_id')['video_title'].to_dict()
[print(i) for i in meta_dict.items()]

#create header & group columns
meta_df["header"] = ['Showcase', 'Recent_released', 'Most_viewed', 'Recent_released', 'Showcase', 
               'Showcase', 'Showcase', 'Most_viewed', 'Recent_released',
               'Most_viewed', 'Recent_released', 'Most_viewed']
meta_df["group"] = ['아이브', '플레이브', '라이즈', '아이브', '이세돌', '플레이브', '라이즈', '이세돌','라이즈', '아이브', '이세돌', '플레이브' ]
meta_df["hg"] =   meta_df["group"] + " | "  + meta_df["header"]
meta_df[['video_title', 'header', 'group', "hg"]]
```

```python
# 조회수 출시 순으로 정렬
inst1 = meta_df[['video_title', 'view_count', 'published_at', 
                 'header', 'group', 'hg']].sort_values(by='published_at')
inst1["published_at"] = pd.to_datetime(inst1["published_at"])
inst1["published_at_TXT"] = inst1["published_at"].dt.year.astype(str) + '/' + inst1["published_at"].dt.month.astype(str)



# 조회수는 누적이기 때문에 공개 순으로 스케일링 해야 할 수도 있음.
## 근데 쇼케이스가 조회수가 젤 낮아서 안해도 될 듯?
```

```python
# Visualization : 총 조회수 (출시일자 순)
import plotly as px
pd.options.plotting.backend = "plotly"
inst1.plot(x='hg', y='view_count', color = 'header', text = 'published_at_TXT', 
           title='총 조회수 Sorted by 공개일',
           kind='bar')
```

```python
# User은 조회 -> 좋아요 -> 댓글 순으로 engage됨
## 조회 -> 좋아요
## 조회 -> 댓글 이기 때문에 fan-engagement를 두 개를 조합해서 확인 할 수 있음

inst2 = meta_df[['video_title', 'view_count', 'like_count', 'commentCount', 'published_at',
                 'header', 'group', 'hg']].sort_values(by='published_at')

#조회수 대비 좋아요 비율
inst2['like_view_ratio'] = inst2['like_count'] / inst2['view_count']

# max & min
max_idx = inst2['like_view_ratio'].idxmax()
min_idx = inst2['like_view_ratio'].idxmin()
print("조회수 대비 좋아요 비율 최고 영상 : ", inst2.loc[max_idx, 'video_title'], inst2.loc[max_idx, 'like_view_ratio'])
print("조회수 대비 좋아요 비율 최저 영상 : ", inst2.loc[min_idx, 'video_title'], inst2.loc[min_idx, 'like_view_ratio'])
print('\n')

#조회수 대비 댓글 비율 
inst2['comment_view_ratio'] = inst2['commentCount'] / inst2['view_count']

max_idx = inst2['like_view_ratio'].idxmax()
min_idx = inst2['comment_view_ratio'].idxmin()
print("조회수 대비 댓글 비율 최고 영상 : ", inst2.loc[max_idx, 'video_title'], inst2.loc[max_idx, 'comment_view_ratio'])
print("조회수 대비 댓글 비율 최저 영상 : ", inst2.loc[min_idx, 'video_title'], inst2.loc[min_idx, 'comment_view_ratio'])
print('\n')
inst2
```

```python
# Visualization : 팬 인게이지먼트 시각화 (조회수 대비 댓글 순) (좋댓구알)

# 상관관계 확인
print("두 지표의 상관계수 :", inst2.corr(numeric_only=True)['comment_view_ratio']['like_view_ratio'])

pd.options.plotting.backend = "plotly"
inst2.plot.bar(x='hg', y='like_view_ratio', color = 'header', 
           title='조회수 대비 좋아요 비율',
           barmode='stack')


```

```python
pd.options.plotting.backend = "plotly"
inst2.plot.bar(x='hg', y='comment_view_ratio', color = 'header', 
           title='조회수 대비 댓글 비율',
           barmode='stack')

# showcase가 높기는 하나, 큰 차이가 벌어지지는 않음. 하지만 조회수 자체의 영향을 받지 않는 것을 보아서, 팬덤의 충성도가 반영된 수치로 보임.
## 두 지표가 의미하는 바가 다르지 않으니 하나만 선정해서 분석 지표로 사용하는 것이 좋을것 같다.
### 우리의 경우에 응용 분석이 댓글로 이어지기 때문에 comment_view_ratio를 상수곱 해서 사용하도록 합시다.
#### 1만 조회수 당 댓글 수 : mill_per_comment = comment_view_ratio * 1000000
```

```python
# 변수 생성
# meta_df['like_view_ratio'] = meta_df['like_count'] / meta_df['view_count']

meta_df['comment_view_ratio'] = meta_df['commentCount'] / meta_df['view_count']

meta_df['mill_per_comment'] = (meta_df['commentCount'] / (meta_df['view_count'] / 1000000)).round(2)
print(meta_df.shape)
meta_df[['video_title', 'view_count', 'commentCount', 'comment_view_ratio', 'mill_per_comment']]
```

### Boy vs Girls
- Boy : 플레이브, RIZZ
- Girl : 이세계아이돌, 아이브

```python
# 그룹 별 인게이지먼트 차이
inst3 = meta_df.copy()
gender_dic = {'플레이브': '남', '라이즈': '남', '이세돌': '여', '아이브': '여'}
inst3['gender'] = inst3['group'].map(gender_dic)

print(inst3.groupby("group")["mill_per_comment"].mean())
print("\n")
      
# 남여 그룹 별 인게이지먼트 차이
print(inst3.groupby("gender")["mill_per_comment"].mean())
```

```python
# visualization : 
inst3.groupby("group")["mill_per_comment"].mean().plot(kind='bar')
```

### Virtual Vs Real
- virtual : 이세계아이돌, 플레이브
- Real : IVE, RIIZE 

```python
inst4 = meta_df.copy()
virtual_dic = {'플레이브': 'Virtual', '라이즈': 'Real', '이세돌': 'Virtual', '아이브': 'Real'}
inst4['virtual'] = inst4['group'].map(virtual_dic)

print(inst4.groupby("virtual")["mill_per_comment"].mean())
```
