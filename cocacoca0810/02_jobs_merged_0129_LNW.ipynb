 ## 3) 중복값 확인

```python
# url이 같을 경우
# 같은 회사의 같은 직무, 같은 시간대에 올라온 공고의 효용 확인
```

```python
df['duplicate_url'] = df['url'].duplicated(keep='first')
```

```python
df['duplicate_url'].value_counts()
```

```python
df.loc[df['url'].duplicated(keep=False),['company', 'title', 'location', 'scrape_date', 'scrape_time', 'url']].sort_values('url')
```

```python
# 회사, 직무명, 날짜시간, 장소가 중복되는 컬럼 생성
df["duplicate_meta"] = df.duplicated(subset=["company", "title", "scrape_time", "scrape_date", "location"],keep='first')
```

```python
df
```

```python
# df["duplicate_meta"].value_counts()
df["duplicate_meta"].value_counts()
```

```python
# duplicate_meta의 True값
df_dup = df[df["duplicate_meta"]]
```

```python
# 링크 대조를 위한 정렬
df_dup.sort_values(by="company")
```

```python
pd.crosstab(df['duplicate_url'], df["duplicate_meta"])
```

```python
def dup_type(row):
    if row['duplicate_url'] and row['duplicate_meta']:
        return 'strict_dup'
    elif row['duplicate_url']:
        return 'url_dup'
    elif row['duplicate_meta']:
        return 'meta_dup'
    else:
        return 'unique'

df['duplicate_type'] = df.apply(dup_type, axis=1)
```

```python
df['duplicate_type'].value_counts()
# meta - url = meta_dup
```

```python
df.loc[:, "duplicate_url":"duplicate_type"]
```

```python
def remove_meta_and_strict_dup(df):
    remove_types = ['meta_dup', 'strict_dup']
    return df[~df['duplicate_type'].isin(remove_types)].copy()
```

```python
df_clean = remove_meta_and_strict_dup(df)
```

```python
df_clean["duplicate_meta"].value_counts()
```

```python
df_clean["duplicate_url"].value_counts()
```

```python
df_clean.info()
```

```python
df_clean_2 = df_clean.drop(columns=["duplicate_meta","duplicate_type"])
```

```python
df_clean_2.info()
```

## 4) 불필요한 행 제거

```python
# "indeed.com" 데이터 없애기
def remove_urls_containing(df_clean_2, keyword):
    return df_clean_2[~df_clean_2['url'].str.contains(keyword, case=False, na=False)].copy()
```

```python
df_new = remove_urls_containing(df_clean_2, 'indeed.com')
```
