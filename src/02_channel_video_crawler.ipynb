```python
all_channel_videos = []

for channel in channels:
    group_label = channel["group_label"]
    channel_label = channel["channel_label"]
    member_label = channel["member_label"]
    handle = channel["handle"]

    print(f"\n=== Fetching videos for {channel_label} ===")

    # 1) handle -> channelId
    channel_id = get_channel_id_from_handle(handle)

    # 2) channelId -> uploads playlistId
    uploads_pid = get_uploads_playlist_id(channel_id)

    # 3) uploads playlist -> ÏòÅÏÉÅ Î™©Î°ù
    videos = get_videos_from_uploads_playlist(
        uploads_playlist_id=uploads_pid,
        channel_label=channel_label,
        channel_handle=handle,
    )

    print(f"{channel_label} videos collected: {len(videos)}")

    # 4) Î©îÌÉÄÎç∞Ïù¥ÌÑ∞ Î≥¥Í∞ï (üî• ÌïµÏã¨ Ï∂îÍ∞Ä Î∂ÄÎ∂Ñ)
    for v in videos:
        v["group_label"] = group_label
        v["member_label"] = member_label

    # 4) Ï†ÑÏ≤¥ Î¶¨Ïä§Ìä∏Ïóê Î≥ëÌï©
    all_channel_videos.extend(videos)

print(f"\nTOTAL videos collected: {len(all_channel_videos)}")
```

```python
df_videos = pd.DataFrame(all_channel_videos)

df_videos[["group_label", "channel_label", "member_label"]].drop_duplicates()
```

```python
BASE_DIR = Path("..")        # src Í∏∞Ï§Ä
DATA_DIR = BASE_DIR / "data"
OUTPUT_PATH = DATA_DIR / "channel_all_videos.csv"

df_videos.to_csv(OUTPUT_PATH, index=False)

print(f"Saved to {OUTPUT_PATH}")
```
