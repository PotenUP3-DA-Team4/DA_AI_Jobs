```python
# í™˜ê²½ ë³€ìˆ˜ ë¡œë“œ (.env)
from dotenv import load_dotenv
load_dotenv()

# í‘œì¤€ ë¼ì´ë¸ŒëŸ¬ë¦¬
import pandas as pd

# íŒŒì¼ ì €ì¥ì„ ìœ„í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬
from pathlib import Path

# ì±„ë„ ê´€ë ¨ ëª¨ë“ˆ
from channel import (
    get_channel_id_from_handle,
    get_uploads_playlist_id,
    get_videos_from_uploads_playlist,
)
```

```python
channels = [
    {"label": "ISEGYE_IDOL", "handle": "isegyeidol_ofcl"},
    {"label": "PLAVE", "handle": "plave_official"},
    {"label": "IVE", "handle": "IVEstarship"},
    {"label": "RIIZE", "handle": "RIIZE_official"},
]
```

```python
all_channel_videos = []

for channel in channels:
    label = channel["label"]
    handle = channel["handle"]

    print(f"\n=== Fetching videos for {label} ===")

    # 1) handle -> channelId
    channel_id = get_channel_id_from_handle(handle)

    # 2) channelId -> uploads playlistId
    uploads_pid = get_uploads_playlist_id(channel_id)

    # 3) uploads playlist -> ì˜ìƒ ëª©ë¡
    videos = get_videos_from_uploads_playlist(
        uploads_playlist_id=uploads_pid,
        channel_label=label,
        channel_handle=handle,
    )

    print(f"{label} videos collected: {len(videos)}")

    # 4) ì „ì²´ ë¦¬ìŠ¤íŠ¸ì— ë³‘í•©
    all_channel_videos.extend(videos)

print(f"\nTOTAL videos collected: {len(all_channel_videos)}")
```

```python
df_videos = pd.DataFrame(all_channel_videos)

df_videos.head(), df_videos.shape
```

```python
BASE_DIR = Path("..")        # src ê¸°ì¤€
DATA_DIR = BASE_DIR / "data"
OUTPUT_PATH = DATA_DIR / "channel_all_videos.csv"

df_videos.to_csv(OUTPUT_PATH, index=False)

print(f"Saved to {OUTPUT_PATH}")
```
