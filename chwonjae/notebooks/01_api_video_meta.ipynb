```python
# ì˜ìƒ ë§í¬ì—ì„œ ì˜ìƒ ID ê°€ì ¸ì˜¤ê¸°
import requests
from dotenv import load_dotenv
from urllib.parse import urlparse, parse_qs
import os

load_dotenv()

def extract_video_id(url: str) -> str | None:
    parsed = urlparse(url)

    # youtu.be í˜•íƒœ
    if parsed.netloc == "youtu.be":
        return parsed.path.lstrip("/")

    # youtube.com/watch í˜•íƒœ
    query = parse_qs(parsed.query)
    return query.get("v", [None])[0]    
```

```python

url = "https://www.googleapis.com/youtube/v3/videos"
video_ids = [extract_video_id(u) for u in video_urls]
data_list = []
for i in video_ids:
    params = {
        
        "part": "snippet,statistics",
        "id": i,
        "key": os.getenv("YOUTUBE_API_KEY")
    }

    response = requests.get(url, params=params)
    data = response.json()
    print(response)
    data_list.append(data)

print(data_list)
```

```python
def fetch_comments(video_id, max_comments=200):
    comments = []
    page_token = None
    
    while True:
        params = {
            "part": "snippet",
            "videoId": video_id,
            "maxResults": 100,        # API ìµœëŒ€
            "order": "time",          # ìµœì‹ ìˆœ
            "textFormat": "plainText",
            "key": os.getenv("YOUTUBE_API_KEY")
        }

        if page_token:
            params["pageToken"] = page_token

        resp = requests.get(
            "https://www.googleapis.com/youtube/v3/commentThreads",
            params=params
        )
        print("STATUS CODE:", resp.status_code)
        print("RAW RESPONSE:", resp.text)

        data = resp.json()
      
        # â— ëŒ“ê¸€ ë¹„í™œì„±í™” / ì˜¤ë¥˜ ì²˜ë¦¬
        if "error" in data:
            print(f"[WARN] comments disabled or error for video_id={video_id}")
            break

        for item in data.get("items", []):
            snippet = item.get("snippet", {})
            top = snippet.get("topLevelComment", {})
            c_snippet = top.get("snippet", {})

            comments.append({
                "video_id": video_id,
                "comment_id": top.get("id"),
                "comment_text": c_snippet.get("textDisplay"),
                "like_count": c_snippet.get("likeCount"),
                "published_at": c_snippet.get("publishedAt")
            })

            if len(comments) >= max_comments:
                return comments

        page_token = data.get("nextPageToken")
        if not page_token:
            break

        time.sleep(0.1)  # quota & ì•ˆì •ì„±

    return comments
```

```python
def fetch_all_comments(video_id, sleep_sec=0.2):
    comments = []
    page_token = None
    page_count = 0

    while True:
        params = {
            "part": "snippet",
            "videoId": video_id,
            "maxResults": 100,
            "order": "time",
            "textFormat": "plainText",
            "key": os.getenv("YOUTUBE_API_KEY")
        }

        if page_token:
            params["pageToken"] = page_token

        resp = requests.get(
            "https://www.googleapis.com/youtube/v3/commentThreads",
            params=params
        )

        data = resp.json()

        # ëŒ“ê¸€ ë¹„í™œì„±í™” / ì˜¤ë¥˜
        if "error" in data:
            print(f"[WARN] skip video_id={video_id}")
            break

        items = data.get("items", [])
        if not items:
            break

        for item in items:
            top = item["snippet"]["topLevelComment"]
            snippet = top["snippet"]

            comments.append({
                "video_id": video_id,
                "comment_id": top["id"],
                "comment_text": snippet["textDisplay"],
                "like_count": snippet["likeCount"],
                "published_at": snippet["publishedAt"]
            })

        page_count += 1
        print(f"page {page_count} | total comments: {len(comments)}")

        page_token = data.get("nextPageToken")
        if not page_token:
            break

        time.sleep(sleep_sec)

    return comments
```

```python
all_comments = []

for i, video in enumerate(video_list, 1):
    video_id = video["video_id"]
    title = video.get("video_title")

    print(f"\n[{i}/{len(video_list)}] Fetching comments for: {title}")

    comments = fetch_all_comments(video_id)

    print(f"â†’ collected {len(comments)} comments")

    all_comments.extend(comments)
```
